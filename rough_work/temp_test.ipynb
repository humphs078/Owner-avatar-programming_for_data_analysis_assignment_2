```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nasa_temp = pd.read_csv('https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt', skiprows=5, header=None, sep = ' ', 
                        skipinitialspace=True, engine='python', names=['year', 'temp_anomaly', 'lowness'])
```

```python
nasa_temp.drop(['lowness'], axis=1, inplace=True)
```

```python
nasa_temp['yr_bp'] = 1950 - nasa_temp['year']
```

```python
nasa_temp = nasa_temp.reindex(columns=['year', 'yr_bp', 'temp_anomaly'])
```

```python
nasa_temp = nasa_temp.sort_values('year', ascending=False)
```

```python
nasa_temp
```

```python
moberg_temp = pd.read_excel('/home/humphs01/Documents/ATU/programming_for_data_analysis_assignment_2/datasets/historic/temperature_dataset.xlsx', sheet_name='2,000 yr',  names=['year', 'yr_bp', 'temp_anomaly', 'x', 'y', 'z'])
```

```python
moberg_temp.drop(['x', 'y', 'z'],axis=1, inplace=True)
```

```python
moberg_temp.drop(moberg_temp.index[0:100], axis = 0, inplace = True)
```

```python
moberg_temp
```

```python
clark_temp = pd.read_excel('/home/humphs01/Documents/ATU/programming_for_data_analysis_assignment_2/datasets/historic/temperature_dataset.xlsx', sheet_name='20,000 yr', names=['yr_bp', 'temp_anomaly', 'x', 'y', 'z'])
```

```python
def temp_year(sample):
    sample.drop(['x', 'y', 'z'],axis=1, inplace=True)
    sample['year'] = 1950 - sample['yr_bp']
    return sample
```

```python
temp_year(clark_temp)
```

```python
clark_temp = clark_temp.reindex(columns=['year', 'yr_bp', 'temp_anomaly'])
```

```python
clark_temp.drop(clark_temp.index[0:19], axis = 0, inplace = True)
```

```python
clark_temp
```

```python
shakun_temp = pd.read_excel('/home/humphs01/Documents/ATU/programming_for_data_analysis_assignment_2/datasets/historic/temperature_dataset.xlsx', sheet_name='800,000 yr', names=['yr_bp', 'temp_anomaly', 'x', 'y', 'z'])
```

```python
temp_year(shakun_temp)
```

```python
shakun_temp = shakun_temp.reindex(columns=['year', 'yr_bp', 'temp_anomaly'])
```

```python
shakun_temp.drop(shakun_temp.index[0:7], axis = 0, inplace = True)
```

```python
shakun_temp
```

```python
frames_temp = [nasa_temp, moberg_temp, clark_temp, shakun_temp]

full_temp_data = pd.concat(frames_temp, ignore_index = True)
```

```python
full_temp_data
```

```python
full_temp_data['years_from_present'] = 2023 - full_temp_data['year']
```

```python
full_temp_data
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(full_temp_data['years_from_present'], full_temp_data['temp_anomaly'])
```
